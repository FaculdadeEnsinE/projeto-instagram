```python

# Quantidade de postagens por ano, com o mais curtido e mais comentado
df["ANO"] = pd.to_datetime(df["timestamp"]).dt.year
postagens_por_ano = (
    # agrupamento por ano
    df.groupby("ANO")
      # o agg faz com que seja possível calcular várias estatísticas ao mesmo tempo
      .agg(
          # quantidade de postagens por ano
          QuantidadePost=("id", "count"),
          # post mais/max curtido por ano
          PostMaisCurtido=("like_count", "max"),
          # post mais/max comentado por ano
          PostMaisComentado=("comments_count", "max")
      )
      # resetar o index para transformar o ano em uma coluna normal
      .reset_index()
      # ordenar por ano
      .sort_values("ANO", ascending=False)
)
display(postagens_por_ano)
```

```python
import requests
from PIL import Image
from io import BytesIO

# URL da media
media_url = df['media_url'][0]
media_type = df['media_type'][0]

# URL da imagem
media_url  = df[df['media_type'] == 'IMAGE']['media_url'].iloc[10]
media_type = df[df['media_type'] == 'IMAGE']['media_type'].iloc[10]

if media_type != "IMAGE":
    print("A mídia não é uma imagem. Tipo:", media_type)
else:
    # Faz a requisição HTTP
    response = requests.get(media_url)

    # Verifica se deu certo
    if response.status_code == 200:
        # Converte o conteúdo em objeto de imagem
        img = Image.open(BytesIO(response.content))
        img.show()  # abre no visualizador padrão
    else:
        print("Erro ao baixar imagem:", response.status_code)
```

```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt

con = duckdb.connect("duckdb/instagram.duckdb", read_only=True)
```

```python
# Total de posts
query_total = """
SELECT COUNT(*) AS total_posts
FROM medias
"""

df_total = con.execute(query_total).df()
df_total
```

```python
# os melhores horários
query_horas_all = """
SELECT
    EXTRACT('hour' FROM timestamp) AS hora,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio,
    ROUND(MEDIAN(like_count + comments_count), 2) AS engajamento_mediano,
    SUM(like_count + comments_count) AS engajamento_total
FROM medias
GROUP BY 1
ORDER BY engajamento_medio DESC, qtd_posts DESC
"""

df_horas_all = con.execute(query_horas_all).df()
df_horas_all
```

```python
# Horários com amostra confiável
query_horas = """
SELECT
    EXTRACT('hour' FROM timestamp) AS hora,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio,
    ROUND(MEDIAN(like_count + comments_count), 2) AS engajamento_mediano,
    SUM(like_count + comments_count) AS engajamento_total
FROM medias
GROUP BY 1
HAVING COUNT(*) >= 10
ORDER BY engajamento_medio DESC, qtd_posts DESC
"""

df_horas = con.execute(query_horas).df()
df_horas
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_horas["hora"], df_horas["engajamento_medio"])
plt.xlabel("Hora")
plt.ylabel("Engajamento médio")
plt.title("Melhores horários para postar")
plt.show()
```

```python
# os melhores dias da semana
query_dias = """
SELECT
    CASE EXTRACT('dow' FROM timestamp)
        WHEN 0 THEN 'Domingo'
        WHEN 1 THEN 'Segunda'
        WHEN 2 THEN 'Terça'
        WHEN 3 THEN 'Quarta'
        WHEN 4 THEN 'Quinta'
        WHEN 5 THEN 'Sexta'
        WHEN 6 THEN 'Sábado'
    END AS dia_semana,
    EXTRACT('dow' FROM timestamp) AS ordem_dia,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio,
    ROUND(MEDIAN(like_count + comments_count), 2) AS engajamento_mediano,
    SUM(like_count + comments_count) AS engajamento_total
FROM medias
GROUP BY 1, 2
ORDER BY engajamento_medio DESC, qtd_posts DESC
"""

df_dias = con.execute(query_dias).df()
df_dias
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_dias["dia_semana"], df_dias["engajamento_medio"])
plt.xticks(rotation=45)
plt.xlabel("Dia da semana")
plt.ylabel("Engajamento médio")
plt.title("Engajamento por dia da semana")
plt.show()
```

```python
# o melhor formato de post
query_tipos = """
SELECT
    media_type,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio,
    ROUND(MEDIAN(like_count + comments_count), 2) AS engajamento_mediano,
    ROUND(AVG(like_count), 2) AS likes_medios,
    ROUND(AVG(comments_count), 2) AS comentarios_medios
FROM medias
GROUP BY 1
ORDER BY engajamento_medio DESC
"""

df_tipos = con.execute(query_tipos).df()
df_tipos
```

```python
plt.figure(figsize=(8,5))
plt.bar(df_tipos["media_type"], df_tipos["engajamento_medio"])
plt.xlabel("Tipo de mídia")
plt.ylabel("Engajamento médio")
plt.title("Desempenho por tipo de post")
plt.show()
```

```python
# quais posts puxaram a média para cima
query_top_posts = """
SELECT
    timestamp,
    media_type,
    like_count,
    comments_count,
    (like_count + comments_count) AS engajamento,
    LEFT(caption, 180) AS legenda
FROM medias
ORDER BY engajamento DESC
LIMIT 20
"""

df_top_posts = con.execute(query_top_posts).df()
df_top_posts
```

```python
# as palavras mais usadas
query_palavras_freq = """
WITH tokens AS (
    SELECT
        UNNEST(REGEXP_EXTRACT_ALL(LOWER(COALESCE(caption, '')), '[[:alpha:]À-ÿ0-9_#@]+')) AS token
    FROM medias
)
SELECT
    REPLACE(token, '#', '') AS palavra,
    COUNT(*) AS frequencia
FROM tokens
WHERE LENGTH(REPLACE(token, '#', '')) >= 4
GROUP BY 1
ORDER BY frequencia DESC
LIMIT 40
"""

df_palavras_freq = con.execute(query_palavras_freq).df()
df_palavras_freq
```

```python
# quais palavras ficaram acima da média do perfil
query_palavras_eng = """
WITH base AS (
    SELECT
        id,
        like_count + comments_count AS engajamento,
        LOWER(COALESCE(caption, '')) AS legenda
    FROM medias
),
media_geral AS (
    SELECT AVG(engajamento) AS media_perfil
    FROM base
),
tokens AS (
    SELECT
        id,
        engajamento,
        UNNEST(REGEXP_EXTRACT_ALL(legenda, '[[:alpha:]À-ÿ0-9_#@]+')) AS token
    FROM base
),
tokens_limpos AS (
    SELECT
        id,
        engajamento,
        REPLACE(token, '#', '') AS palavra
    FROM tokens
    WHERE LENGTH(REPLACE(token, '#', '')) >= 4
)
SELECT
    palavra,
    COUNT(*) AS frequencia,
    ROUND(AVG(engajamento), 2) AS engajamento_medio_palavra,
    ROUND(AVG(engajamento) - (SELECT media_perfil FROM media_geral), 2) AS acima_da_media_do_perfil
FROM tokens_limpos
GROUP BY 1
HAVING COUNT(*) >= 10
ORDER BY acima_da_media_do_perfil DESC, frequencia DESC
LIMIT 40
"""

df_palavras_eng = con.execute(query_palavras_eng).df()
df_palavras_eng
```

```python
#o melhor horário por formato
query_hora_tipo = """
SELECT
    EXTRACT('hour' FROM timestamp) AS hora,
    media_type,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio,
    ROUND(MEDIAN(like_count + comments_count), 2) AS engajamento_mediano
FROM medias
GROUP BY 1, 2
HAVING COUNT(*) >= 5
ORDER BY engajamento_medio DESC, qtd_posts DESC
"""

df_hora_tipo = con.execute(query_hora_tipo).df()
df_hora_tipo
```

```python
# Provar temas específicos
query_temas = """
SELECT
    'forro' AS tema,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio
FROM medias
WHERE LOWER(caption) LIKE '%forró%' OR LOWER(caption) LIKE '%forro%'

UNION ALL

SELECT
    'sorteio' AS tema,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio
FROM medias
WHERE LOWER(caption) LIKE '%sorteio%'

UNION ALL

SELECT
    'ingresso' AS tema,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio
FROM medias
WHERE LOWER(caption) LIKE '%ingresso%' OR LOWER(caption) LIKE '%ingressos%'

UNION ALL

SELECT
    'workshop' AS tema,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio
FROM medias
WHERE LOWER(caption) LIKE '%workshop%'

UNION ALL

SELECT
    'agradecimento' AS tema,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio
FROM medias
WHERE LOWER(caption) LIKE '%obrigado%' OR LOWER(caption) LIKE '%obrigada%'
"""

df_temas = con.execute(query_temas).df()
df_temas
```

```python
# se CTA melhora desempenho (CTA significa “Call to Action”, ou em português, “chamada para ação”.)
# É qualquer frase ou elemento no seu conteúdo que orienta o público a fazer algo específico.
query_cta = """
SELECT
    CASE
        WHEN LOWER(caption) LIKE '%comenta%'
          OR LOWER(caption) LIKE '%marca%'
          OR LOWER(caption) LIKE '%salva%'
          OR LOWER(caption) LIKE '%envia%'
        THEN 'Com CTA'
        ELSE 'Sem CTA'
    END AS grupo_cta,
    COUNT(*) AS qtd_posts,
    ROUND(AVG(like_count + comments_count), 2) AS engajamento_medio,
    ROUND(MEDIAN(like_count + comments_count), 2) AS engajamento_mediano
FROM medias
GROUP BY 1
ORDER BY engajamento_medio DESC
"""

df_cta = con.execute(query_cta).df()
df_cta
```

Análise Final: 

A análise realizada considerou um conjunto de 1.263 publicações do perfil, avaliando indicadores como curtidas, comentários, tipo de mídia, horário e dia das postagens, além do conteúdo textual das legendas. O objetivo foi identificar padrões consistentes de comportamento do público e, a partir disso, orientar decisões estratégicas para aumento de alcance e engajamento.

No que diz respeito aos horários de postagem, os dados indicam que o desempenho mais consistente ocorre no período da tarde, especialmente entre 14h e 17h, com destaque para 17h, 15h e 14h. Embora o horário das 09h tenha apresentado uma média elevada de engajamento, essa evidência se baseia em um volume muito pequeno de publicações, o que reduz sua confiabilidade para tomada de decisão. Dessa forma, a recomendação é priorizar a faixa da tarde, que apresenta resultados mais estáveis e sustentados por uma amostra significativa de dados. Esse comportamento sugere que o público do perfil tende a estar mais ativo após o almoço e no período de transição entre atividades do dia.

Ao analisar os dias da semana, observa-se que o melhor desempenho ocorre em dias úteis, com destaque para quarta-feira, seguida por segunda e quinta-feira. Em contrapartida, o domingo apresenta o pior resultado em termos de engajamento médio. Esse padrão indica que o público está mais receptivo ao conteúdo durante a rotina semanal, enquanto nos finais de semana, especialmente no domingo, há uma redução na atenção e na interação, possivelmente devido a outras prioridades e hábitos de consumo diferentes.

Em relação ao formato das publicações, os dados mostram que o carrossel é o tipo de conteúdo com maior engajamento médio, seguido pelos vídeos. As imagens estáticas apresentam desempenho inferior. Esse resultado reforça a lógica de que conteúdos que demandam mais tempo de atenção do usuário, como deslizar imagens ou assistir a um vídeo, tendem a ser mais valorizados pelo algoritmo e geram maior envolvimento do público. Portanto, há uma oportunidade clara de otimização ao priorizar esses formatos mais dinâmicos.

A análise dos posts com maior engajamento revela um padrão importante: conteúdos relacionados a sorteios, convites para eventos, datas especiais, agradecimentos e mensagens com apelo emocional estão entre os que mais performam. O ponto em comum entre esses conteúdos é a presença de estímulos à ação ou à conexão emocional. Em outras palavras, quanto mais o conteúdo convida o público a participar, reagir ou se identificar, maior tende a ser o engajamento gerado.

No campo das legendas, foi possível identificar uma forte repetição de termos relacionados ao nicho do perfil, como “forró”, “djkalango” e variações associadas à comunidade. Isso demonstra um posicionamento claro e consistente, o que é positivo para construção de identidade e reconhecimento. No entanto, a repetição excessiva pode limitar o alcance para novos públicos. Em paralelo, palavras associadas a convite, participação e benefício, como “sorteio”, “ingressos”, “workshop”, “participar”, “você” e “obrigado”,  aparecem com desempenho acima da média do perfil, indicando que conteúdos mais interativos e diretos tendem a gerar melhores resultados.

Outro ponto relevante é o impacto das chamadas para ação (CTA). A análise mostra que publicações que incluem orientações explícitas ao usuário, como comentar, marcar alguém ou compartilhar, apresentam, em média, desempenho superior às que não possuem esse direcionamento. Isso reforça a importância de conduzir o comportamento do público de forma clara, reduzindo a fricção na interação.

Além disso, o cruzamento entre horário e tipo de conteúdo evidencia que o desempenho não depende apenas de quando postar, mas também do que é postado em cada momento. Isso indica que a estratégia ideal deve considerar a combinação entre formato, contexto e horário, e não apenas um único fator isolado.

De forma geral, os dados revelam que o perfil já possui um ativo importante: uma identidade bem definida e um público engajado dentro de um nicho específico. No entanto, existem oportunidades claras de crescimento por meio de ajustes estratégicos. Entre as principais recomendações estão: priorizar postagens entre 14h e 17h, concentrar conteúdos mais relevantes nos dias úteis, especialmente quarta, segunda e quinta, investir com maior frequência em carrosséis e vídeos, utilizar chamadas para ação de forma consistente e diversificar levemente o conteúdo para ampliar o alcance sem perder a identidade.
