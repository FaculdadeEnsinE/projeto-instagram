# Conectar Duckdb to DataFrame Pandas

```python
# Use a variável path_dir que você já definiu + o nome do arquivo
database_path = path_dir + 'duckdb/instagram.duckdb'

print(database_path)

# Conectar ao banco de dados usando o caminho completo
conn = duckdb.connect(database_path)

# ler a tabela medias do duckdb
df = conn.execute("SELECT * FROM medias").fetchdf()

# desconectar do banco de dados
conn.close()
```

# Consultas SQL convertidas para Dataframa

```python
# Quantidade total de mídias
total_medias = len(df)
display(f"Total de mídias: {total_medias}")
```

```python
# Quantidade de mídias por tipo
medias_por_tipo = df.groupby("media_type").size().reset_index(name="count").sort_values("media_type")
display(medias_por_tipo)
```

```python
# Quantidade total de curtidas
total_likes = df["like_count"].sum()
display(f"Total de curtidas: {total_likes}")
```

```python
# Quantidade total de comentários
total_comments = df["comments_count"].sum()
display(f"Total de comentários: {total_comments}")
```

```python
# Postagens com mais comentários (top 5)
top_comentadas = df.sort_values("comments_count", ascending=False).head(5)
display(top_comentadas[["id", "caption", "comments_count"]])
```

```python
# Postagem mais antiga
mais_antiga = df.sort_values("timestamp", ascending=True).head(1)
display(mais_antiga[["id", "caption", "timestamp"]])
```

```python

# Quantidade de postagens por ano, com o mais curtido e mais comentado
df["ANO"] = pd.to_datetime(df["timestamp"]).dt.year
postagens_por_ano = (
    # agrupamento por ano
    df.groupby("ANO")
      # o agg faz com que seja possível calcular várias estatísticas ao mesmo tempo
      .agg(
          # quantidade de postagens por ano
          QuantidadePost=("id", "count"),
          # post mais/max curtido por ano
          PostMaisCurtido=("like_count", "max"),
          # post mais/max comentado por ano
          PostMaisComentado=("comments_count", "max")
      )
      # resetar o index para transformar o ano em uma coluna normal
      .reset_index()
      # ordenar por ano
      .sort_values("ANO", ascending=False)
)
display(postagens_por_ano)
```

# Abrir imagem pelo link media url

```python
# DataFrame com imagem
df_imagem = df[df["media_type"] == "IMAGE"]
display(df_imagem.head())
```

```python
import requests
from PIL import Image
from io import BytesIO

import random

# Gerar um número inteiro aleatório entre 0 e o número de linhas do dataframe de imagens - 1
random_integer = random.randint(0, df_imagem.shape[0] - 1)

# URL da media / imagem
media_url  = df_imagem['media_url'].iloc[random_integer]
media_type = df_imagem['media_type'].iloc[random_integer]
caption     = df_imagem['caption'].iloc[random_integer]
print("URL da imagem:", media_url)
print("Tipo da mídia:", media_type)
print("Indice da imagem:", random_integer)
print("Legenda:", caption)

if media_type != "IMAGE":
    print("A mídia não é uma imagem. Tipo:", media_type)
else:
    # Faz a requisição HTTP
    response = requests.get(media_url)

    # Verifica se deu certo
    if response.status_code == 200:
        # Converte o conteúdo em objeto de imagem
        img = Image.open(BytesIO(response.content))
        img.show()  # abre no visualizador padrão
    else:
        print("Erro ao baixar imagem:", response.status_code)
```

```python
# 1. Preparar os dados (Média por ano)
df_media = df.groupby('ano')[['like_count', 'comments_count']].mean()

# 2. Criar o gráfico com um estilo limpo
fig, ax = plt.subplots(figsize=(12, 6))

# Definindo cores modernas (Azul Suave e Rosa Pastel)
cores = ['#3498db', '#e74c3c']

# Plotar as barras com bordas arredondadas (opcional) e cores
df_media.plot(kind='bar', ax=ax, color=cores, width=0.8, edgecolor='white', linewidth=1)

# 3. Títulos e Legendas
plt.title("Engajamento Médio Anual: Curtidas vs Comentários", fontsize=16, pad=20, fontweight='bold', color='#2c3e50')
plt.xlabel("Ano de Postagem", fontsize=12, labelpad=10)
plt.ylabel("Média de Interações", fontsize=12, labelpad=10)
plt.xticks(rotation=0) # Anos na horizontal

# Ajustar a legenda para ficar fora do gráfico ou em um local limpo
plt.legend(["Média de Curtidas", "Média de Comentários"], frameon=False, fontsize=10)

# 4. Limpeza Visual (Remover bordas e adicionar grade suave)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(True, linestyle='--', alpha=0.3) # Grade apenas horizontal e clarinha
ax.set_axisbelow(True) # Garante que a grade fique atrás das barras

# 5. Adicionar rótulos de dados (os números em cima das barras)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (0, 9),
                textcoords = 'offset points',
                fontsize=9, color='#7f8c8d')

plt.tight_layout() # Ajusta tudo para não cortar nada
plt.show()
```
