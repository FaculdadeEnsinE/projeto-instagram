```python
# Importa as bibliotecas necessárias
import json

# pandas para manipulação de dados
import pandas as pd

# Glob para ler os arquivos JSON
import glob
```

```python

# Listar os arquivos JSON na pasta
json_files = glob.glob("./json/*.json")

# Ler todos os arquivos JSON e concatenar em um único DataFrame
dfs = []
for file in json_files:
    with open(file, "r", encoding="utf-8") as f:
        dfs.append(pd.json_normalize(json.load(f)))
df_all = pd.concat(dfs, ignore_index=True)

# Converter para Spark DataFrame
spark_df = spark.createDataFrame(df_all)

# Apagar a tabela se existir
spark.sql("DROP TABLE IF EXISTS instagram.bronze.post")

# Criar a tabela com os dados
spark_df.write.format("delta").saveAsTable("instagram.bronze.post")
```
